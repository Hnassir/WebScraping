# function to Get and parse html content from a page

```python
def parser_html(url):

    import requests
    from bs4 import BeautifulSoup
    
    response=requests.get(url)

    soup=BeautifulSoup(response.content,'html.parser')


    return soup


def article_title(soup):
    from bs4 import BeautifulSoup
    
    if bool(soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('h1')):
        title=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('h1')
    else :
        title=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('p')
    
        
    return title.text
    
def paragraphes(soup):
         
    from bs4 import BeautifulSoup

    dict={}
    
    if bool(soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('h2')):
        parag_t=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('h2')
        parag_1=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('p')

    else:
        parag_t=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find_all('div',attrs={"class":'g_heading_rich w-richtext'})[1]
        parag_1=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find_all('div',attrs={'class':'w-richtext'})[2].find('p')



    dict[parag_t.text]=parag_1.text

    for parag in soup.find_all('div',attrs={"class":'casessingle_body_layout u-grid-desktop u-gap-large'}):
        if bool(parag.find('h3')):
            dict[parag.find('h3').text]=parag.find('p').text
        else :
            var=parag.find_all('p')
            dict[var[0].text]=var[1].text
            
    
    return dict


def collect_links(soup,url):

    from bs4 import BeautifulSoup

    link="https://www.easyagile.com"
    lst=[]

    link_h=soup.find('div',attrs={"class":'casessingle_header_content u-column-6 u-vflex-left-center u-gap-small'}).find('a')['href']
    lst.append(link+link_h)

    for parag in soup.find_all('div',attrs={"class":'casessingle_body_layout u-grid-desktop u-gap-large'}):
        lst.append(link+parag.find('a')['href'])

    for links in soup.find('div',attrs={"class":'use_cases_collection_wrap w-dyn-list'}).find_all('div',attrs={'class':'w-dyn-item'}):
        if url != link+links.find('a')['href']:
            lst.append(link+links.find('a')['href'])
    

    return lst
```

```python
def web_scraping(url):
    
    import pandas as pd

    dic={}

    print('wrap all')
    print(url)
    soup=parser_html(url)
    title_h=article_title(soup)
    dict=paragraphes(soup)
    lst=collect_links(soup,url)

    print('head',title_h)
    print('paragraphes',dict)

    dic[title_h]=[var for var in dict]
    dic['paragraph_content']=[var for var in dict.values()]
    
    print('links',lst)
    print(len(lst))
    dic['all_links']=[lst for i in dict]
    
    return pd.DataFrame(dic)
```

```python
page_='https://www.easyagile.com/use-cases/retrospectives'

y='https://www.easyagile.com/use-cases/share-objectives-and-track-progress'
web_page=web_scraping(page_)

web_page

web_page2=web_scraping(y)

web_page2
```
